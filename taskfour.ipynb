```python
# PCA for Handwritten dataset
# Author: Sukriti
from scipy.io import loadmat
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb

# load matlab files
handwritten = loadmat ('USPS.mat')

# find out dataset
handwritten.keys()

# store in matrix
mat = handwritten['A']

# display original last image
oglast = np.reshape(mat[2999], (16, 16))
sb.heatmap(oglast, annot = False)
plt.xlabel("Original Last Image")
plt.show()

# display original second last image
oglast = np.reshape(mat[2998], (16, 16))
sb.heatmap(oglast, annot = False)
plt.xlabel("Original Second Last Image")
plt.show()



######
# PCA transformation for p = 10
######
print("For p=10")
pca = PCA(n_components = 10)
pca.fit(mat)
# transformation 1
trans = pca.transform(mat)
print("Size of p=10 reduced matrix ", trans.shape)
dataframetest = pd.DataFrame(data = trans)

# reconstructed image
inverse_test = pca.inverse_transform(dataframetest)

# p = 10 print after reconstruction
recons = np.reshape(inverse_test[2999], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 10 Last Image')
plt.show()

recons = np.reshape(inverse_test[2998], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 10 Second Last Image')
plt.show()


# p = 10 calculating error
lost = ((mat - inverse_test) ** 2).mean()
print('For p = 10 the reconstruction error is ' +str(lost))


######
# PCA transformation for p = 50
######
print("For p=50")
pca = PCA(n_components = 50)
pca.fit(mat)
# transformation 1
trans = pca.transform(mat)
print("Size of p=50 reduced matrix ", trans.shape)
dataframetest = pd.DataFrame(data = trans)

# reconstructed image
inverse_test = pca.inverse_transform(dataframetest)

# p = 50 print after reconstruction
recons = np.reshape(inverse_test[2999], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 50 Last Image')
plt.show()

recons = np.reshape(inverse_test[2998], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 50 Second Last Image')
plt.show()


# p = 50 calculating error
lost = ((mat - inverse_test) ** 2).mean()
print('For p = 50 the reconstruction error is ' +str(lost))




######
# PCA transformation for p = 100
######
print("For p=100")
pca = PCA(n_components = 100)
pca.fit(mat)
# transformation 1
trans = pca.transform(mat)
print("Size of p=100 reduced matrix ", trans.shape)
dataframetest = pd.DataFrame(data = trans)

# reconstructed image
inverse_test = pca.inverse_transform(dataframetest)

# p = 100 print after reconstruction
recons = np.reshape(inverse_test[2999], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 100 Last Image')
plt.show()

recons = np.reshape(inverse_test[2998], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 100 Second Last Image')
plt.show()


# p = 100 calculating error
lost = ((mat - inverse_test) ** 2).mean()
print('For p = 100 the reconstruction error is ' +str(lost))





######
# PCA transformation for p = 150
######
print("For p=150")
pca = PCA(n_components = 150)
pca.fit(mat)
# transformation 1
trans = pca.transform(mat)
print("Size of p=150 reduced matrix ", trans.shape)
dataframetest = pd.DataFrame(data = trans)

# reconstructed image
inverse_test = pca.inverse_transform(dataframetest)

# p = 150 print after reconstruction
recons = np.reshape(inverse_test[2999], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 150 Last Image')
plt.show()

recons = np.reshape(inverse_test[2998], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 150 Second Last Image')
plt.show()


# p = 150 calculating error
lost = ((mat - inverse_test) ** 2).mean()
print('For p = 150 the reconstruction error is ' +str(lost))



######
# PCA transformation for p = 200
######
print("For p=200")
pca = PCA(n_components = 200)
pca.fit(mat)
# transformation 1
trans = pca.transform(mat)
print("Size of p=200 reduced matrix ", trans.shape)
dataframetest = pd.DataFrame(data = trans)

# reconstructed image
inverse_test = pca.inverse_transform(dataframetest)

# p = 150 print after reconstruction
recons = np.reshape(inverse_test[2999], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 200 Last Image')
plt.show()

recons = np.reshape(inverse_test[2998], (16,16))
sb.heatmap(recons, annot = False)
plt.xlabel('p = 200 Second Last Image')
plt.show()


# p = 200 calculating error
lost = ((mat - inverse_test) ** 2).mean()
print('For p = 200 the reconstruction error is ' +str(lost))







```
